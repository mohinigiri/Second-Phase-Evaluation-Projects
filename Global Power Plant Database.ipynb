```python
def itna_data_kam_kyu_hai(df,threshold):
    total_rows = len(df)
    null_threshold = threshold*total_rows/100
    
    data_kam_hai = []
    
    for column in df.columns:
        null_count = df[column].isnull().sum()
        null_percentage = (null_count/total_rows)*100
        if null_percentage > threshold:
            data_kam_hai.append(column)
            
    return data_kam_hai
            
        
    
    
```
